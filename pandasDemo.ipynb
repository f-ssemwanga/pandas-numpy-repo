### High Level Statistics about the data

```python
df.describe() # quick data metrics


```

```python
#Sorting data
#df.sort_values('Name') # sorted by name in ascending order
df.sort_values('Name', ascending=False) # sorted by name in descending order

#sorting on more than one column
df.sort_values(['Type 1', 'HP'], ascending=False)
#sorting in different directions
df.sort_values(['Type 1', 'HP'], ascending=[1,0]) #Type 1 is ascending and HP is descending
```

### Making changes to the data



```python
df.head(5)
#add all the stats and create a total column
df['Total'] = df['HP']+df['Attack'] +  df['Defense']+df['Sp. Atk']+  df['Sp. Def']+df['Speed']
df.head(5)
```

```python
#drop a specific column

df = df.drop(columns=['Total'])
df.head(5)

#add total column in a different way
df['Total'] = df.iloc[:, 4:10].sum(axis=1) #: means all rows and columns [4:9], axis=1 is adding horizontally and 0 vertically
df.head(5)

#move the position of the total column
cols = list(df.columns.values) # puts all the columns in a list
df = df[cols[0:4] + [cols[-1]]+ cols[4:12]]  #note the single colum cols[-1] is sorrounded in []
df.head(5)

```

### Exporting the modified data to the desired file format

```python
#Saving the updated data to a csv

df.to_csv('modified.csv', index=False) # index = False does not create an extra index
df.to_excel('modifiedExcel.xlsx', index=False) # saving to excel
df.to_csv('modified.txt', index=False,sep='\n') # saving to text / tab separated document
```

### Filtering data

```python
df.loc[(df['Type 1']== "Grass") & (df['Type 2']=='Poison')] #filtering on two columns 
#can use the | for the OR operator

new_df = df.loc[(df['Type 1']== "Grass") & (df['Type 2']=='Poison') & (df['HP']>70)]

#resetting the index
new_df = new_df.reset_index(drop=True) #removes the old index and creates new one
new_df.to_csv('filtered_csv')
```

```python
#filtering all that contain specific terms
df.loc[df['Name'].str.contains('Mega')][0:5]
```

```python
#filtering all that exclude a  specific terms
df.loc[~df['Name'].str.contains('Mega')][0:5]
```

```python
#filtering all that contain a regex
import regex
import re

df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)][0:5] #filter out Grass or fire, flag for ignore case
```

```python
df.loc[df['Name'].str.contains('^pi[a-z*]', flags=re.I, regex=True)][0:5] #all in Name column starting with pi
```
