Example the following [People Favourites CSV File](PeoplesFavourites-1.csv)to determine how the date needs to be cleaned. This file does contain duplicates and it does contain inconsistencies. Selecting appropriate structures and techniques perform the following tasks to prepare the data:

- Anonymize the data by removing the customer names (first and last name) replacing with a unique identifier.

```python
import pandas as pd


df = pd.read_csv("PeoplesFavourites-1.csv", delimiter=',')
df.head(10)
```

##### Remove any customer data that has missing data in any of the following first name, last name and email.

- I will use the **dropna()** method with the ** subset** parameter 

```python
#removing rows with missing data in first name, last name or Email
df.dropna(subset=['First Name', 'Last Name', 'Email'], inplace = True)
df.head(5)
```

##### Anonymize the data by removing the customer names (first and last name) replacing with a unique identifier.

```python

import uuid

#anonymise customer's names by generating a unique ID
df['Unique ID'] = [str(uuid.uuid4()) for _ in range(len(df))] #create a uuid for each record
#df['Unique Identifier'] = df.apply(lambda _: str(uuid.uuid4()), axis=1)  # alternative approach

#remove the unwanted columns
col_to_drop = ["First Name","Last Name" ]
existing_columns = [col for col in col_to_drop if col in df.columns]
df.drop(columns=col_to_drop, inplace=True)
df.head(5)

```

##### A number of different symbols (including space) have been used for null values, select and apply a consist ant approach to their representation.

- To apply a consistent approach I have opted  to use the **'replace()** method to replace all the different symbols used for null values with the standard NaN representation
- The following are the symbols I will be replacing:
"none", "NaN, "0", "N/A", "", " " (space)

```python
import numpy as np
#create a list of null symbols
null_symbols = ['none', 'NaN', '0', 'N/A', '', ' ']
df.replace(null_symbols, np.nan, inplace=True)
df.tail(15)
```
